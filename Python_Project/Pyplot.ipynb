Pyplot Tutorial

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("indices")
plt.title("myplot")
plt.show()
```

```python
plt.plot([1,3,5,7],[2,4,6,8])
plt.ylabel("Numbers")
plt.xlabel("indices")
plt.title("myplot2")
plt.grid()
plt.show()
```

```python
plt.plot([1,3,5,7],[2,4,6,8],'ro')  #ro is used to draw thw red color dots
plt.ylabel("Numbers")
plt.xlabel("indices")
plt.title("myplot3")
plt.grid()
plt.show()
```

```python
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t*2,'b--',label='^2') #blue dashed line
plt.plot(t,t**3,'go:',label='^2.2') #green dotted line with
plt.plot(t,t**2.5,'r*-',label='^2.5') #red dash dot line with triangle marker
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y,'r^-',linewidth=2)  #line width is used to increase the thickness of line
plt.show()
```

```python
x1=[1,2,3,4,5]
y1=[2,3,5,7,11]
x2=[1,2,3,4,5]
y2=[1,4,6,8,10]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0) #in python style
plt.setp(lines[1],'color','g',linewidth=2.0)  #matlab style
plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r-')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1,2,3')
plt.show()
```
