```python
weather_data = [('2024-01-01', 30,6, 'Sunny'),
                ('2024-01-02', 28,5, 'Cloudy'),
                ('2024-01-03', 25,4, 'Rainy'),
                ('2024-01-04', 32,7, 'Sunny'),
                ('2024-01-05', 27,3, 'Windy'),
                ('2024-01-06', 29,4, 'Rainy')
                ]
df = pd.DataFrame(weather_data, columns=['Date', 'Temperature', 'Windspeed', 'Condition'])
print(df)
```
